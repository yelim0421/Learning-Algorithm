**거스름돈**(그리디 알고리즘)

> 당신은 음식점의 계산을 도와주는 점원이다. 카운터에는 거스름돈으로 사용할 500원, 100원, 50원, 10원짜리 동전이 무한히 존재한다고 가정하자. 손님에게 거슬러 줘야 할 돈이 N원일 때 거슬러 줘야 할 동전의 최소 개수를 구하여라. 단, 거슬러 줘야 할 돈 N원은 항상 10의 배수이다.




```python
n=1260
count=0

coin_types=[500,100,50,10]      #큰 화폐의 순서대로 차례대로 확인

for coin in coin_types:
  count+=n//coin
  n%=coin

print(count)
```

**큰 수의 법칙**(그리디 알고리즘)

> 숫자 배열에서 M번 더하여 가장 큰 수를 만드는데, 특정한 수를 K번 초과해서 연속 더할 수 없다.

> 입력 조건: 첫째 줄에 배열의 크기N, M, K이 주어지며 공백으로 구분한다. 둘째 줄에 N개의 자연수가 주어진다. 입력으로 주어지는 K는 항상 M보다 작거나 같다. 


> 출력 조건: 첫째 줄에 큰 수의 법칙에 따라 더해진 답을 출력한다





```python
#가장 큰 수를 K번 더하고 두번째로 큰 수를 한번 더하는 연산을 반복-> 제일큰수, 두번째로 큰 수만 필요함

n,m,k=map(int,input().split())        #N,M,K를 공백으로 구분하여 입력받기
data=list(map(int,input().split()))   #N개의 수를 공백으로 구분하여 입력받기
data.sort()                           #데이터 정렬하기(오름차순!)

first=data[n-1]                       #제일 큰 수
second=data[n-2]

result=0

while True:
  for i in range(k):
    if m==0:
      break
    result+=first
    m-=1
  if m==0:
    break
  result+=second
  m-=1

print(result)
```

```python
#위 문제를 보면 6+6+6+5 가 계속 반복된다. 이 수열의 길이는 K+1이 된다.
#M(더하는 번)/(K+1)=(K+1)가 반복되는 횟수
#여기에 K를 다시 곱해주면 제일 큰수가 더해지는 횟수가 된다.

#만약 M(더하는 번)/(K+1)=이 나누어 떨어지지 않는다면 나머지만큼 큰 수를 더해줘야 한다.
#(M(더하는 번)/(K+1))*K+M%(K+1)=큰 수가 더해지는 횟수

n,m,k=map(int,input().split())        #N,M,K를 공백으로 구분하여 입력받기
data=list(map(int,input().split()))   #N개의 수를 공백으로 구분하여 입력받기
data.sort()                           #데이터 정렬하기(오름차순!)

first=data[n-1]                       #제일 큰 수
second=data[n-2]

count=int(m/(k+1))*k
count+=m%(k+1)

result=0
result+=(count)*first
result+=(m-count)*second
print(result)
```

**숫자 카드 게임**

> 여러개의 숫자 카드 중에서 가장 높은 숫자가 쓰인 카드 한 장을 뽑는 게임이다.

> 첫째 줄에 숫자 카드들이 놓인 행의 개수N과 열의 개수M이 공백을 기준으로 자연수로 주어진다. 둘째 줄부터 N개의 줄에 걸쳐 각 카드에 적힌 숫자가 주어진다.

> 첫째 줄에 게임의 룰에 맞게 선택한 카드에 적힌 숫자를 출력한다.






```python
#각 행마다 가장 작은 수를 찾은 다음에 그 수 중에서 가장 큰 수를 찾는 것
#min()함수를 이용하기

n,m=map(int,input().split())

result=0

#한 줄씩 입력받아 확인
for i in range(n):
  data=list(map(int,input().split()))
  #현재 줄에서 '가장 작은 수' 찾기
  min_value=min(data)
  result=max(result,min_value)
#print(result)
```

```python
#2중 반복문 구조 이용하기

n,m=map(int,input().split())

result=0
for i in range(n):
  data=list(map(int,input().split()))
  min_value=10001                       #일단 제일 큰 수로 정해놓고
  for a in data:
    min_value=min(min_value,a)
  result=max(result,min_value)
print(result)
```

> 첫째 줄에 N이 1이 될때까지 1번 혹은 2번 과정을 수행해야 하는 최솟값을 출력한다.
